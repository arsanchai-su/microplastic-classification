#Inferance example

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
import torch
efficientnet_model = torch.load('/content/efficientnet_microplastics.pt',map_location=torch.device(device))
resnetv2_50_model = torch.load('/content/resnetv2_50_microplastics.pt',map_location=torch.device(device))
resnetv2_101_model = torch.load('/content/resnetv2_101_microplastics.pt',map_location=torch.device(device))
mobilenetv3_model = torch.load('/content/mobilenetv3_microplastics.pt',map_location=torch.device(device))
inception_v3_model = torch.load('/content/inception_v3_microplastics.pt',map_location=torch.device(device))
```

```python
class_names = {0:"Fiber",1:"Film",2:"Fragment",3:"Pellet"}
```

```python
from typing import List, Tuple
from PIL import Image
import torch
import torchvision.transforms as T
import matplotlib.pyplot as plt
import numpy as np

def pred_class_and_show_image(model: torch.nn.Module,
                              image_path: str,
                              class_names: List[str],
                              image_size: Tuple[int, int] = (224, 224),
                              device: torch.device = torch.device('cpu')):
    img = Image.open(image_path)

    image_transform = T.Compose([
        T.Resize(image_size),
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])

    model.to(device)

    model.eval()
    with torch.inference_mode():
        transformed_image = image_transform(img).unsqueeze(dim=0)
        target_image_pred = model(transformed_image.to(device))
        target_image_pred_probs = torch.softmax(target_image_pred, dim=1)
        target_image_pred_label = torch.argmax(target_image_pred_probs, dim=1).cpu().numpy()

        classname = class_names[target_image_pred_label[0]]
        prob = target_image_pred_probs.max().cpu().numpy()

        # Display the image along with class name and probability
        plt.imshow(np.array(img))
        plt.title(f'Class: {classname}, Probability: {prob:.4f}')
        plt.axis('off')
        plt.show()
```

```python
pred_class_and_show_image(efficientnet_model,"/content/example_fiber.jpg", class_names)
pred_class_and_show_image(efficientnet_model,"/content/example_film.jpg", class_names)
pred_class_and_show_image(efficientnet_model,"/content/example_fragment.jpg", class_names)
pred_class_and_show_image(efficientnet_model,"/content/example_pellet.jpg", class_names)
```

```python
pred_class_and_show_image(resnetv2_50_model,"/content/example_fiber.jpg", class_names)
pred_class_and_show_image(resnetv2_50_model,"/content/example_film.jpg", class_names)
pred_class_and_show_image(resnetv2_50_model,"/content/example_fragment.jpg", class_names)
pred_class_and_show_image(resnetv2_50_model,"/content/example_pellet.jpg", class_names)
```
